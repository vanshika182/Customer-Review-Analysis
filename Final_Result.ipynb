```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/LDA_PROJECT_FOLDER/lda_topic_distributions_with_scores.csv")

# Show the first few rows
df.head()
```

```python
# Step 1: Manual topic name mapping
topic_names = {
    0: "Texture",
    1: "Packaging",
    2: "Beverage",
    3: "Sauces",
    4: "Coffee",
    5: "Meals",
    6: "Kids",
    7: "Snacks",
    8: "Pets",
    9: "Delivery",
    10: "Complaints",
    11: "Tea",
    12: "Reviews",
    13: "Gluten",
    14: "Chocolate",
    15: "Cooking",
    16: "Spicy",
    17: "Energy",
    18: "Organic",
    19: "Taste",
    20: "K-Cups",
    21: "Protein",
    22: "Pricing"
}

# Step 2: Map to DataFrame
df['Topic_Name'] = df['Dominant_Topic'].map(topic_names)
```

```python
df
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from ast import literal_eval

# Download VADER
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
```

```python
# Perform VADER sentiment analysis
def get_sentiment(lemmatized_text):
    score = sia.polarity_scores(lemmatized_text)
    compound = score['compound']
    if compound >= 0.05:
        return 'Positive'
    elif compound <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment'] = df['lemmatized_text'].apply(get_sentiment)

# Compute sentiment percentage per topic
topic_sentiment_pct = pd.crosstab(
    df['Dominant_Topic'],
    df['Sentiment'],
    normalize='index'
) * 100

# Rename columns and reset index
topic_sentiment_pct = topic_sentiment_pct.rename(columns={
    'Positive': 'Positive_%',
    'Neutral': 'Neutral_%',
    'Negative': 'Negative_%'
}).reset_index()

# Merge sentiment percentages with main DataFrame
df = df.merge(topic_sentiment_pct, on='Dominant_Topic', how='left')

# Final output table
final_df = df[['lemmatized_text', 'Dominant_Topic', 'Topic_Name', 'Positive_%', 'Neutral_%', 'Negative_%']]
final_df.head(10)
```

```python
final_df = final_df.reset_index(drop=True)
final_df['Review_ID'] = final_df.index
```

```python
grouped = final_df.groupby('Dominant_Topic').agg({
    'Review_ID': list,
    'Positive_%': 'first',
    'Neutral_%': 'first',
    'Negative_%': 'first',
    'Topic_Name': 'first'  # include topic name here
}).reset_index()
```

```python
for _, row in grouped.iterrows():
    topic = row['Dominant_Topic']
    topic_name = row['Topic_Name']
    review_ids = row['Review_ID']
    pos = round(row['Positive_%'], 2)
    neu = round(row['Neutral_%'], 2)
    neg = round(row['Negative_%'], 2)

    print(f"Review {', '.join(map(str, review_ids))} belong to Topic {topic} ({topic_name})")
    print(f"Topic Name: {topic_name}")
    print(f"Sentiment: Positive: {pos}%, Neutral: {neu}%, Negative: {neg}%\n")

```

```python
with open("review_topic_sentiment_summary1.txt", "w") as f:
    for _, row in grouped.iterrows():
        topic = row['Dominant_Topic']
        review_ids = row['Review_ID']
        pos = round(row['Positive_%'], 2)
        neu = round(row['Neutral_%'], 2)
        neg = round(row['Negative_%'], 2)

        f.write(f"Review {', '.join(map(str, review_ids))} belong to Topic {topic}\n")
        f.write(f"Topic {topic} Sentiment: Positive: {pos}%, Neutral: {neu}%, Negative: {neg}%\n\n")
```
